```python
import pandas as pd

import math


AA=pd.DataFrame(columns = ['criteria','weight','ratingA','ratingB','scoreA','scoreB'])

AA['criteria'] = ['profitability', 'improvedProductivity','costSavings','investmentRequired','Customers','Revenue','Organization']

AA['weight'] = [0.2,0.1,0.15,0.1,0.2,0.1,0.15]

AA['ratingA'] = [8,2,4,4,3,5,6]

AA['ratingB'] = [7,3,4,3,8,5,7]


for index, row in AA.iterrows():
    AA['scoreA'][index]=row['ratingA']*row['weight']
    AA['scoreB'][index]=row['ratingB']*row['weight']

totalScoreA=0
totalScoreB=0
for index, row in AA.iterrows():
  totalScoreA+=row['scoreA']
  totalScoreB+=row['scoreB']

print(AA)
```

```python
import numpy as np
import matplotlib.pyplot as plt

axisLocations = np.linspace(start=0, stop=2 * np.pi, num =len(AA['ratingA']), endpoint=False)
axisLocations = np.concatenate((axisLocations,[axisLocations[0]]))

ratingsA=AA['ratingA']
ratingsB=AA['ratingB']
ratingsACircular=np.concatenate((ratingsA,[ratingsA[0]]))
ratingsBCircular=np.concatenate((ratingsB,[ratingsB[0]]))

plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(axisLocations, ratingsACircular, 'o-', linewidth=2, label='Economic Impact')
plt.plot(axisLocations, ratingsBCircular, 'd-', linewidth=2, label='Risk')
plt.title('Strategy Goals Prioritization Map', size=20)
lines, labels = plt.thetagrids(np.degrees(axisLocations), labels=AA['criteria'])
plt.legend()
plt.show()
```
